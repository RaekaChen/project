## <H3>2.4. Treating missing values </H3>

```python
def missing_data(df):
    total = df.isnull().sum().sort_values(ascending = False)
    Percentage = (df.isnull().sum()/df.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, Percentage], axis=1, keys=['Total', 'Percentage']).reset_index()

missing_data(df)
```

```python
# Remove rows from with 0 video views
df.drop(df.loc[df.video_views == 0].index, inplace=True)

# Validate the results        
filtered_df = df[df['video_views'] == 0]
filtered_df
```
